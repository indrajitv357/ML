```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

## Setting the global parameters to ideal font sizes and types, can also be done locally 
matplotlib.rcParams.update({'font.size': 18, 'font.family': 'STIXGeneral', 'mathtext.fontset': 'stix'})


x = np.linspace(0,5,10)
y = x ** 2



### Manually handling axis sub plots and positioning
### Aspect ratio and figure plot size can be controlled when creating the figure object

fig = plt.figure(figsize=(8,6), dpi= 340) ### Figure object
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y,'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('Test')


### Can use subplot to automatically handle plot positioning

fig, axes = plt.subplots(nrows = 1, ncols = 2, figsize = (8,6), dpi = 340)

for ax in axes:
    ax.plot(x, y, 'r')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('title')
    
fig.tight_layout ## Ensures that the grib positioning of each graph is done correctly so as to avoid axis and label overlap


fig, ax1 = plt.subplots(dpi = 360)

## The r converts the following string into a raw text format, so that the $<>$ encapsualtion can be used for LATEX format
ax1.scatter(x,x**3,label = r"$y = \alpha^3$") ## scatter plot instead of a regular line
ax1.plot(x,x**4,label = r"$y = \alpha^4$")
ax1.legend(loc=0) ## Automatic legend creation and update -> loc =0 (determines optimal location)
ax1.set_xlabel(r"$\alpha$")
ax1.set_ylabel("Y")
ax1.set_title("LATEX format")
ax1.axis('tight') ## Automatically configures the axis ranges to get tightly fitted axis for the range of data being plotted 
ax1.grid(color='b', alpha=0.5, linestyle='dashed', linewidth=0.5)


## Fill between type plot, to highlight difference in results (also consider twinx and twiny type displays)
fig, ax2 = plt.subplots(dpi = 360)
ax2.fill_between(x, x**2, x**3, color="green", alpha=0.5)
ax2.grid(color = 'b', alpha = 0.5, linestyle = 'dashed', linewidth = 0.5)

### pie plots
fracs = [30, 15, 45, 10]
colors = ['b', 'g', 'r', 'w']
# 1st argument of pie should be data
fig, ax = plt.subplots(figsize=(6, 6))  # make the plot square
pie = ax.pie(fracs, colors=colors, explode=(0, 0, 0.05, 0), shadow=True,
             labels=['A', 'B', 'C', 'D'])



```
